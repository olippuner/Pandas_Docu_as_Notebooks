# Getting started

## Installation

> pandas is part of the [Anaconda](https://docs.continuum.io/anaconda/)
distribution and can be installed with Anaconda or Miniconda:

```bash
conda install pandas
```


—

pandas can be installed via pip from [PyPI](https://pypi.org/project/pandas).

```bash
pip install pandas
```


Installing a specific version? Installing from source? Check the advanced
installation page.

> Refer to this corresponding page on online documentation. It has additional information about installation.
